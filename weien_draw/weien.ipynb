```python
import matplotlib.pyplot as plt
from venn import venn

# 定义六个集合
set1 = {1, 2, 3}
set2 = {2, 3, 4}
set3 = {3, 4, 5}
set4 = {4, 5, 6}
set5 = {5, 6, 7}
set6 = {6, 7, 8}

# 创建一个字典，将集合名称与集合内容关联
sets = {
    'Set 1': set1,
    'Set 2': set2,
    'Set 3': set3,
    'Set 4': set4,
    'Set 5': set5,
    'Set 6': set6
}

# 绘制韦恩图
venn(sets)

# 显示图形
plt.title("Venn Diagram of Six Sets")
plt.show()
```
